```python
# Sistema de Gestión de Biblioteca con Estructuras Lineales (Versión Colab)

# Estructuras
libros = []
cola_prestamos = []
pila_devoluciones = []

# Funciones
def agregar_libro():
    id_libro = input("Ingrese ID del libro: ")
    titulo = input("Ingrese título: ")
    autor = input("Ingrese autor: ")
    anio = input("Ingrese año de publicación: ")
    try:
        anio = int(anio)
        if any(libro['ID'] == id_libro for libro in libros):
            print("❌ ID de libro ya se encuentra registrado, por favor validar y volver a ingresar.")
        else:
            libros.append({
                'ID': id_libro,
                'Título': titulo,
                'Autor': autor,
                'Año': anio
            })
            print("✅ Libro agregado correctamente.")
    except ValueError:
        print("❌ Año inválido. Debe ser un número entero.")

def listar_libros():
    if not libros:
        print("No hay libros registrados.")
    else:
        print("\nLista de libros:")
        for i, libro in enumerate(libros, 1):
            print(f"{i}. {libro['Título']} (ID: {libro['ID']})")

def prestar_libro():
    id_libro = input("Ingrese ID del libro a prestar: ")
    libro_prestado = next((libro for libro in libros if libro['ID'] == id_libro), None)
    if libro_prestado:
        cola_prestamos.append(libro_prestado)
        print(f"✅ Libro '{libro_prestado['Título']}' prestado.")
    else:
        print("❌ Libro no encontrado.")

def devolver_libro():
    id_libro = input("Ingrese ID del libro a devolver: ")
    libro_devuelto = next((libro for libro in libros if libro['ID'] == id_libro), None)
    if libro_devuelto:
        pila_devoluciones.append(libro_devuelto)
        print(f"✅ Libro '{libro_devuelto['Título']}' devuelto.")
    else:
        print("❌ Libro no encontrado.")

def eliminar_libro():
    id_libro = input("Ingrese ID del libro a eliminar: ")
    global libros
    libros = [libro for libro in libros if libro['ID'] != id_libro]
    print(f"✅ Libro con ID '{id_libro}' eliminado.")

def mostrar_prestamos():
    if not cola_prestamos:
        print("No hay libros prestados.")
    else:
        print("\nLibros prestados:")
        for libro in cola_prestamos:
            print(f"- {libro['Título']} (ID: {libro['ID']})")

def mostrar_devoluciones():
    if not pila_devoluciones:
        print("No hay libros devueltos.")
    else:
        print("\nLibros devueltos:")
        for libro in pila_devoluciones:
            print(f"- {libro['Título']} (ID: {libro['ID']})")

# Menú
def menu_biblioteca():
    while True:
        print("\n Sistema de Gestión de Biblioteca ")
        print("1. Agregar Libro")
        print("2. Listar Libros")
        print("3. Prestar Libro")
        print("4. Devolver Libro")
        print("5. Eliminar Libro")
        print("6. Ver Préstamos")
        print("7. Ver Devoluciones")
        print("8. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            agregar_libro()
        elif opcion == '2':
            listar_libros()
        elif opcion == '3':
            prestar_libro()
        elif opcion == '4':
            devolver_libro()
        elif opcion == '5':
            eliminar_libro()
        elif opcion == '6':
            mostrar_prestamos()
        elif opcion == '7':
            mostrar_devoluciones()
        elif opcion == '8':
            print("Saliendo del sistema.")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

menu_biblioteca()
```
